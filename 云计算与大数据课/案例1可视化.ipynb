```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('./data/TrainingSet.csv')
test_data = pd.read_csv('./data/TestSet.csv')
```

```python
temp = train_data.groupby('EndDay')['QuantitySold'].sum()
days = temp.index.values
success = temp.values
plt.figure(figsize=(8, 5))
plt.rcParams['font.sans-serif']=['SimHei']
plt.bar(days, success, color='blue', label='TrainSet')
plt.xlabel('日期')
plt.ylabel('成功数量')
plt.title('TrainSet每日拍卖成功数量图')
plt.legend()
plt.savefig('./image/1-1.png')
plt.show()
```

```python
temp = test_data.groupby('EndDay')['QuantitySold'].sum()
days = temp.index.values
success = temp.values
plt.figure(figsize=(8, 5))
plt.rcParams['font.sans-serif']=['SimHei']
plt.bar(days, success, color='green', label='TrainSet')
plt.xlabel('日期')
plt.ylabel('成功数量')
plt.title('TestSet每日拍卖成功数量图')
plt.legend()
plt.savefig('./image/1-2.png')
plt.show()
```

```python
temp1 = train_data.groupby('EndDay')['QuantitySold'].sum()
temp2 = train_data.groupby('EndDay')['QuantitySold'].count()
days = temp2.index.values
rate = temp1.values / temp2.values
plt.figure(figsize=(8, 5))
plt.bar(days, rate, color='yellow', label='TrainSet')
plt.xlabel('日期')
plt.ylabel('成功率')
plt.title('TrainSet每日拍卖成功率')
plt.legend()
plt.savefig('./image/1-3.png')
plt.show()
```

```python
temp1 = test_data.groupby('EndDay')['QuantitySold'].sum()
temp2 = test_data.groupby('EndDay')['QuantitySold'].count()
days = temp2.index.values
rate = temp1.values / temp2.values
plt.figure(figsize=(8, 5))
plt.bar(days, rate, color='red', label='TrainSet')
plt.xlabel('日期')
plt.ylabel('成功率')
plt.title('TestSet每日拍卖成功率')
plt.legend()
plt.savefig('./image/1-4.png')
plt.show()
```
